```python
import pandas as pd
from datetime import datetime
import json 

def addDashtoNodeName(nodename):
    nodename=nodename.split('node')
    return('node_'+nodename[1])

def columnFormat(columnName):
    columnName=columnName.lower()
    if columnName[0:4]=='node':
        columnName=addDashtoNodeName(columnName)
    return columnName

def monthMapper(month):
    date = datetime(2000, month, 1)  # We need a dateobject with the proper month
    return date.strftime('%B').lower()

```

```python
voltage_files={'voltage_a':'./IEEE123_data/output/nodes_voltage_A.csv','voltage_b':'./IEEE123_data/output/nodes_voltage_B.csv','voltage_c':'./IEEE123_data/output/nodes_voltage_C.csv'}
voltage_dfs={}
voltage_dfs_flattened={}
for voltage_name,filename in voltage_files.items():
    voltage_dfs[voltage_name]=pd.read_csv(filename,skiprows=8,parse_dates=['# timestamp'])

for voltage_name,df in voltage_dfs.items():
    df.columns = [columnFormat(col_name) for col_name in df.columns]
    df['month'] = df['# timestamp'].map(lambda x: monthMapper(x.month))
    df['hour_of_day'] = df['# timestamp'].map(lambda x: x.hour)
    df['day_of_month'] = df['# timestamp'].map(lambda x: x.day)

    ### Get all the node names
    nodeNames=[]
    for columnName in df.columns:
        if columnName[0:4]=='node':
            nodeNames.append(columnName)

    df=pd.melt(df, id_vars=['month','hour_of_day','day_of_month'],var_name='node',value_vars=nodeNames, value_name=voltage_name+'_magnitude')
    df[voltage_name+'_magnitude']=df[voltage_name+'_magnitude'].map(lambda x:abs(complex(x.strip('+0i'))))
    voltage_dfs_flattened[voltage_name]=df
voltageData = reduce(lambda left,right: pd.merge(left,right,on=['month', 'day_of_month', 'hour_of_day','node']), voltage_dfs_flattened.values())
dercamData=pd.read_csv('../results/test_results.csv')
final_down_selected=voltageData[voltageData['month']=='march'][voltageData['day_of_month']==2][["month","hour_of_day","node","voltage_b_magnitude","voltage_c_magnitude" ,"voltage_a_magnitude"]]
merge_test=dercamData.append(final_down_selected)

merge_test.drop('Unnamed: 0',axis=1).to_csv('merge_test.csv')




```

```python
from datetime import datetime
def isWeekday(month,day_of_month,year=2017):
    weekday=datetime.strptime(month+":"+str(day_of_month)+":"+str(year), '%B:%d:%Y').weekday()
    if weekday in [5,6]:
        return 'Weekend'
    else:
        return 'Weekday'
    

voltageData['Weekday'] = voltageData.apply(lambda row: isWeekday(row['month'], row['day_of_month'],2017), axis=1)
```

```python
heatmap={}
voltages={'voltage_A':'voltage_a_magnitude',
          'voltage_B':'voltage_b_magnitude',
          'voltage_C':'voltage_c_magnitude',
         }
for key,value in voltages.iteritems():
    heatmap_hourly_dict={}
    for hour_of_day in range(0,24):
        heatmap_list=[]
        heatmap_hourly={}
        for elem in set(list(merge_test['node'])):
            try:
                heatmap_list.append({'lat':node_coordinates[elem][1],'lng':node_coordinates[elem][0],
                                     'count':merge_test[(merge_test['node']==elem) & (merge_test['hour_of_day']==hour_of_day)][value].values[0]})
            except:
                continue
        heatmap_hourly['max']=2530
        heatmap_hourly['min']=2170
        heatmap_hourly['data']=heatmap_list
        heatmap_hourly_dict[hour_of_day]=heatmap_hourly
    heatmap[key]=heatmap_hourly_dict
```
